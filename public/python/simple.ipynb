```python
from bokeh.palettes import Spectral4
from bokeh.plotting import figure, show

p = figure(x_axis_type="datetime")
for gasoline, color in zip(['Gazole', 'SP98', 'SP95'], Spectral4):
    data = duckdb.sql(f"""
        select *
        from daily
        where
            extract(year from date) >= 2023
            and prix_nom = '{gasoline}'
        order by date
    """)
    data_df = data.df()
    p.line(data_df['date'], data_df['median_price'], line_width=2, color=color, legend_label=gasoline)

show(p)
```
